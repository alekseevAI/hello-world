```python
import numpy as np
import matplotlib.pyplot as plt
data = np.loadtxt('populations.txt', dtype = {'names': ('year', 'hare', 'lynx', 'carrot'), 'formats': ('i4', 'f4', 'f4', 'f4')})
```

```python
plt.plot(data['year'], data['hare'])
plt.plot(data['year'], data['lynx'])
plt.plot(data['year'], data['carrot'])
plt.xlabel('Year')
plt.ylabel('Population')
plt.legend(('hare', 'lynx', 'carrot'))
plt.show()
```

```python
plt.bar (data['year'], data['lynx'] / (data['lynx'] + data['hare']) * 100)
plt.xlabel('Year')
plt.ylabel('Percent')
plt.title('lynx')
plt.show()

plt.bar (data['year'], 100 - (data['lynx'] / (data['lynx'] + data['hare']) * 100))
plt.xlabel('Year')
plt.ylabel('Percent')
plt.title('hare')
plt.show()
```

```python
print('lynx: {}, {}'.format(np.mean(data['lynx']), np.std(data['lynx'])))
print('hare: {}, {}'.format(np.mean(data['hare']), np.std(data['hare'])))
print('carrot: {}, {}'.format(np.mean(data['carrot']), np.std(data['carrot'])))

plt.hist(data['lynx'], 10)
plt.axvline(x = np.mean(data['lynx']), color = 'red')
plt.title('lynx')
plt.show()

plt.hist(data['hare'], 10)
plt.axvline(x = np.mean(data['hare']), color = 'red')
plt.title('hare')
plt.show()

plt.hist(data['carrot'], 10)
plt.axvline(x = np.mean(data['carrot']), color = 'red')
plt.title('carrot')
plt.show()
```

```python
print('lynx: {}'.format(data['year'][data['lynx'].argmax()]))
print('hare: {}'.format(data['year'][data['hare'].argmax()]))
print('carrot: {}'.format(data['year'][data['carrot'].argmax()]))

plt.plot(data['year'], data['hare'], 'r')
plt.plot(data['year'], data['lynx'], 'g')
plt.plot(data['year'], data['carrot'], 'b')
plt.axvline(x=data['year'][data['hare'].argmax()], color='r', linestyle='-')
plt.axvline(x=data['year'][data['lynx'].argmax()], color='g', linestyle='-')
plt.axvline(x=data['year'][data['carrot'].argmax()], color='b', linestyle='-')
plt.xlabel('Year')
plt.ylabel('Population')
plt.legend(('hare', 'lynx', 'carrot'))
plt.show()
```

```python
a = np.sign(data['hare'] - data['lynx'])
b = list(map(lambda x: 'H' if x == 1 else 'L', a))
plt.pie([b.count('H'),b.count('L')], labels=['H', 'L'], autopct='%1.f')
plt.show()
```

```python
print('hare: {}'.format(data['year'][np.where(data['hare'] > 50000)]))
print('lynx: {}'.format(data['year'][np.where(data['lynx'] > 50000)]))
print('carrot: {}'.format(data['year'][np.where(data['carrot'] > 50000)]))

plt.plot(data['year'], data['hare'], 'r')
plt.plot(data['year'], data['lynx'], 'g')
plt.plot(data['year'], data['carrot'], 'b')

plt.vlines(data['year'][np.where(data['hare'] > 50000)], 0, 90000, color = 'r', linestyle='--', alpha=0.5)
plt.vlines(data['year'][np.where(data['lynx'] > 50000)], 0, 90000, color = 'g', linestyle='--',  alpha=0.5)
plt.vlines(data['year'][np.where(data['carrot'] > 50000)], 0, 90000, color = 'b', linestyle='--',  alpha=0.5)

plt.xlabel('Year')
plt.ylabel('Population')
plt.legend(('hare', 'lynx', 'carrot'))
plt.axis([1900, 1920, 0, 90000])
plt.show()
```

```python
print('hare: {}'.format(data['year'][np.argsort(data['hare'])[0:2]]))
print('lynx: {}'.format(data['year'][np.argsort(data['lynx'])[0:2]]))
print('Tcarrot: {}'.format(data['year'][np.argsort(data['carrot'])[0:2]]))
```

```python
plt.plot(data['year'][1:], data['hare'][1:] - data['hare'][:20] , 'b')
plt.plot(data['year'][1:], data['lynx'][1:], 'r')

plt.legend(('hare dif', 'lynx',))
plt.axis([1901, 1920, -50000, 70000])
plt.show()

print(np.corrcoef(data['hare'][1:] - data['hare'][:20], data['lynx'][1:])[0][1])
```
